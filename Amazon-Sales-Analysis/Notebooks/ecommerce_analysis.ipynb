```python
import pandas as pd
import sqlite3
```

```python
conn=sqlite3.connect('Amazon_sales.db')
cursor_obj=conn.cursor()
```

```python
#Load the cleaned CSV
cleaned_data=pd.read_csv('cleaned_amazon_sales_report.csv')
```

```python
# Write to SQLite table
cleaned_data.to_sql('amazon_sales_data', conn, if_exists='replace',index=False)
```

```python
#Verify import
new_data=pd.read_sql('''select * from amazon_sales_data ''',conn)
new_data
```

```python
#Total Revenue & Average Order Value
new_data=pd.read_sql("""select round(sum("Total Revenue"),2) as total_revenue,
                     round(avg("Total Revenue"),2) as avg_revenue from 
                     amazon_sales_data """,conn)
new_data
```

```python
# Revenue by category
new_data=pd.read_sql("""select Category, sum("Total Revenue") as total_revenue from 
                     amazon_sales_data
                     group by 1
                     order by 2 desc """,conn)
new_data
```

```python
#Top 5 Revenue by state
new_data=pd.read_sql("""select "ship-state", sum("Total Revenue") as total_revenue from 
                     amazon_sales_data
                     group by 1
                     order by 2 desc """,conn)
new_data
```

```python
#Best-Selling categories
new_data=pd.read_sql("""select Category, sum(Qty) as total_sold from 
                     amazon_sales_data
                     group by 1
                     order by 2 desc """,conn)
new_data
```

```python
# Daily Sales Trend
new_data=pd.read_sql('''select sum("Total Revenue") as total_revenue, strftime('%d', Date) as daily_revenue from amazon_sales_data 
                     group by 2''',conn)
new_data
```
