```python
import pandas as pd
import numpy as np
```

```python
ecom=pd.read_csv("Amazon Sale Report.csv")
ecom
```

```python
ecom.info() #Check for missing values and data types
ecom.describe()
```

```python
#Remove missing values
ecom.dropna(inplace=True)
ecom
```

```python
#Converted date column to datetime format:
ecom['Date'] = pd.to_datetime(ecom['Date'])
ecom
```

```python
#calculate total revenue per order
ecom['Total Revenue']=ecom['Qty']*ecom['Amount']
ecom
```

```python
ecom.drop(['Unnamed: 22'],axis=1, inplace=True)
```
